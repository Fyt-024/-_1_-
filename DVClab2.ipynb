```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 
df = pd.read_csv("merged_common_features_encoded.csv")

# 
corr_matrix = df.corr()

# 
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Features Correlation Heatmap")
plt.show()
```

```python
print(corr_matrix["selling_price"].sort_values(ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
df = pd.read_csv("merged_common_features_encoded.csv")

# 1)  year 
plt.figure()
plt.scatter(df["year"], df["selling_price"], alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.title("Year vs. Selling Price")
plt.show()

# 2)  km_driven 
plt.figure()
plt.scatter(df["km_driven"], df["selling_price"], alpha=0.5)
plt.xlabel("km_driven")
plt.ylabel("Selling Price")
plt.title("km_driven vs. Selling Price")
plt.show()
```

```python
# transmission_encoded vs. selling_price 
plt.figure()
df.boxplot(column="selling_price", by="transmission_encoded", grid=False)
plt.title("Transmission vs. Selling Price")
plt.suptitle("")  # 
plt.xlabel("Transmission Encoded")
plt.ylabel("Selling Price")
plt.show()
```
